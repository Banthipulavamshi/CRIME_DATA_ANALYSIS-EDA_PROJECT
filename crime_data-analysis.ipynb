```python
# hightest crime rate

top_crime_locations = data['AREA NAME'].value_counts().head(10)
top_crime_locations_data = top_crime_locations.reset_index()
top_crime_locations_data.columns = ['AREA NAME', 'Crime Count']

print(top_crime_locations_data)
```

```python
#Least crime rate

crime_counts = data['AREA NAME'].value_counts()
least_crime_areas = crime_counts[crime_counts == crime_counts.min()]

print("Least crime area names:")
print(least_crime_areas)
```

```python
#Minors in the crime
#(ages 10-19)
minors_data = data[(data['Vict Age'] >= 10) & (data['Vict Age'] <= 19)]
print(minors_data)
```

```python
minors_data = data[(data['Vict Age'] >= 15) & (data['Vict Age'] <= 30)]

# Count incidents per crime type
crime_counts = minors_data['Crm Cd Desc'].value_counts().reset_index()
crime_counts.columns = ['Crime Type', 'Count']

# Create a bar plot
plt.figure(figsize=(10, 7))
sns.barplot(data=crime_counts, x='Crime Type', y='Count',)
plt.title('Number of Crimes Involving Minors (Ages 10-19)')
plt.xticks(rotation=45, ha='right')
plt.xlabel('Crime Type')
plt.ylabel('Number of Incidents')
plt.tight_layout()
plt.show()
```
